## Supplier Dict

```python
def supplier_dict_func(path):
    df = pd.read_csv(path)
    ticker = df["ticker"][0]
    name = df["Firm Name"][0]
    supplier_dict = df.drop(columns = ["ticker","Firm Name"]).set_index('Supplier Name').agg(list, axis=1).to_dict()
    arr = np.array([[ticker,name,supplier_dict]])
    df = pd.DataFrame(arr, columns = ["ticker","Firm Name","supplier_dict"])
    return df
```

```python
supplier_df = pd.DataFrame()
for ticker in tickers:
    path = "/Users/minhthao/Desktop/MSDA/Summer 2023/Bloomberg/SPLC Images/CSVs/" +ticker + "/suppliers.csv"
    if "suppliers.csv" in os.listdir("/Users/minhthao/Desktop/MSDA/Summer 2023/Bloomberg/SPLC Images/CSVs/" +
                                     ticker + "/"):
        
        data = supplier_dict_func(path)
        supplier_df = pd.concat([supplier_df,data],axis = 0)
    else:
        print("File from " + ticker + " does not exist")
        pass
```

## Customer Dict

```python
def customer_dict_func(path):
    df = pd.read_csv(path)
    ticker = df["ticker"][0]
    name = df["Firm Name"][0]
    supplier_dict = df.drop(columns = ["ticker","Firm Name"]).set_index('Customer Name').agg(list, axis=1).to_dict()
    arr = np.array([[ticker,name,supplier_dict]])
    df = pd.DataFrame(arr, columns = ["ticker","Firm Name","customer_dict"])
    return df
```

## Competitors Dict
* Drop the first 4 rows
* The fifth rows is the original company and below rows are its competitors
* Strip of row that has Nan and below
* Strip of anything after ticker name
* Create a table - Ticker Name, Company Name, Comp Ticker, Comp Name

```python
def competitor_file(path):
    df = pd.read_excel(path)
    df = df[["EQY_FUND_CRNCY","REL_INDEX"]]
    df = df.drop(df.index[:4], axis=0).reset_index(drop=True)
    
    # Find the anything before the first whitespace
    regex = re.compile(r"^(.*?)\s")
    df["Competitor Ticker"] = df["EQY_FUND_CRNCY"].str.extract(regex)
    row_index = df['EQY_FUND_CRNCY'].isnull().idxmax()
    df = df.iloc[:row_index]
    ticker = df["Competitor Ticker"].loc[0]
    name = df["REL_INDEX"].loc[0]
    df = df.iloc[1:]
    comp_dict = dict(zip(df["Competitor Ticker"], df["REL_INDEX"]))
    arr = np.array([[ticker,name,df.shape[0],comp_dict]])
    return arr
```

```python
def get_file_paths(folder_path):
    file_paths = []
    for root, dirs, files in os.walk(folder_path):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            file_paths.append(file_path)
    return file_paths
```

```python
competitors = get_file_paths("/Users/minhthao/Desktop/MSDA/Summer 2023/Bloomberg/SPLC Images/Competitors")
new_arr = np.empty((0, 4))
for comp in competitors:
    arr = competitor_file(comp)
    new_arr = np.vstack((new_arr, arr))

comp_df = pd.DataFrame(new_arr, columns = ["ticker","Firm Name","comp_num","comp_dict"])
```
